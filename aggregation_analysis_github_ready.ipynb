## Figure 1: Exponent renormalization map

The identity line $R(p)=p$ is shown for reference.  
A stable fixed point appears where the renormalization map intersects the identity.

```python
plt.figure(figsize=(7,5))
plt.plot(p_test, Rs_structured, "o-", label="Structured interference")
plt.plot(p_test, Rs_scrambled, "s--", label="Phase-scrambled control")
plt.plot(p_test, p_test, "k:", label="Identity: R(p)=p")
plt.axvline(2.0, linestyle=":")
plt.axhline(2.0, linestyle=":")
plt.xlabel("Micro exponent p")
plt.ylabel("Renormalized exponent R(p)")
plt.title("Exponent Renormalization Map")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Figure 2: Flow function

The sign of $\beta(p)=R(p)-p$ indicates the renormalization direction.  
A stable fixed point occurs where $\beta(p)$ crosses zero with flow directed toward the crossing from both sides.

```python
flow_structured = Rs_structured - p_test
flow_scrambled = Rs_scrambled - p_test

plt.figure(figsize=(7,5))
plt.plot(p_test, flow_structured, "o-", label=r"Structured $\beta(p)=R(p)-p$")
plt.plot(p_test, flow_scrambled, "s--", label=r"Scrambled $\beta(p)=R(p)-p$")
plt.axhline(0.0, linestyle="--")
plt.axvline(2.0, linestyle=":")
plt.xlabel("p")
plt.ylabel(r"$\beta(p)=R(p)-p$")
plt.title("RG Flow Function for the Exponent")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Fixed-point estimate
